```python
import pandas as pd

with open("NLPAssignment1-sp23-18032024-011322pm.txt" , 'r') as file:
    txt = file.read() 
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
capital_words = []
tokens = word_tokenize(txt)
for token in tokens:
    if token[0].isupper():
        capital_words.append(token)
    print(f'Word : {token}, No of Characters : {len(token)}')

print(f'Start with Capital letters : {capital_words}')
```

```python
from nltk.util import bigrams, trigrams

#Unigrams :
print(tokens)
 #bigrams :
bigram = list(bigrams(tokens))
print(bigram)

#Trigram
trigram = list(trigrams(tokens))
print(trigram)
```

```python
import re
import docx2txt
#docx2txt for document and PyPDF2 for pdf
cv = docx2txt.process("Muheed Ahmed Khan.docx")
```

```python
pattern = r'.*?(bachelor|masters).*?'
match_string = re.search(pattern, cv, re.IGNORECASE)  # for searching both cases
if match_string:
    print(match_string.group())

```

```python
pattern = r'\(\+\d{2}\) \d{3}-\d{7}'
match_string = re.search(pattern, cv)
if match_string:
    print(match_string.group())
```

```python
cv_text = """Last degree: MS(SE)
Experience in years:6
Contact: 0333-2454322
"""

pattern = r'Experience\s+in\s+years:\s*(\d+)'
match_string = re.search(pattern, cv_text , re.IGNORECASE)
if match_string:
    print(match_string.group(1))
```
